# Preparing data

```python
import geopandas as gpd
from shapely.geometry import mapping
```

```python
estados=gpd.read_file('../files/maps/geojson/GeoJSON - Estados.json')
# estados = gpd.read_file('../files/maps/Shapefile - Censo 2010 (Estatal)/inegi_refcenesta_2010.shp')
estados
```

```python
# Cleaning data
estados['nom_ent'] = estados['nom_ent'].replace('Michoacán de Ocampo','Michoacán')
estados['nom_ent'] = estados['nom_ent'].replace('Coahuila de Zaragoza','Coahuila')
estados['nom_ent'] = estados['nom_ent'].replace('Veracruz de Ignacio de la Llave','Veracruz')
estados['nom_ent'] = estados['nom_ent'].replace('Distrito Federal','Ciudad De México')
estados['abrv_edo'] = estados['abrv_edo'].replace('DF','CDMX')
```

```python
df_map = estados.merge(df, right_on='State',left_on='nom_ent',how='inner').drop(columns=['State'])
df_map = df_map.explode(index_parts=True)
df_map
```

```python
df_map2 = df_map.copy()
df_map2['Percentage'] = (df_map2['Percentage'].round(2) * 100).astype(int)
```

```python
from matplotlib_scalebar.scalebar import ScaleBar
from shapely.geometry import Polygon
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import geoplot.crs as gcrs
import mapclassify as mc
import contextily as ctx
import geoplot as gplt
import seaborn as sns
```

```python
# Configurar el estilo de la gráfica
plt.style.use('fivethirtyeight')

# Crear el esquema de colores usando la biblioteca mapclassify
# La función Quantiles permite transformar los datos numéricos a un esquema de clases
# scheme = mc.Quantiles(df_map['Name'].dropna(), k=5)

# Crear la figura y el eje
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# Convertir el GeoDataFrame al sistema de coordenadas EPSG:3857
df_map = df_map.to_crs(epsg=3857)

# Crear el mapa coropletico con los datos del Percentage carencias 
# y los poligonos por municipio
gplt.choropleth(
    df_map,
    hue="Name",
    # scheme=scheme, 
    cmap= 'BuRd',
    linewidth=.1,
    edgecolor='black',
    legend=True,
    ax=ax
)
# Ajustar los límites del eje para que coincidan con los datos proyectados
ax.set_xlim(df_map.total_bounds[0], df_map.total_bounds[2])
ax.set_ylim(df_map.total_bounds[1], df_map.total_bounds[3])

# Añadir el mapa base del proveedor Openstreetmap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik, zoom=6)

# Obtener el tamaño del mapa para calcular la posición de la imagen
# Esto se hace para posicionar la rosa de los vientos
map_width = ax.get_xlim()[1] - ax.get_xlim()[0]
map_height = ax.get_ylim()[1] - ax.get_ylim()[0]

# Definir las coordenadas para la posición de la imagen
img_x = ax.get_xlim()[0] + map_width * 2 / 3  # Aproximadamente un tercio del ancho del mapa desde la izquierda
img_y = ax.get_ylim()[0] + map_height / 2  # Centrado verticalmente en el mapa

# Ajustar el tamaño de la imagen de la rosa de los vientos
img_width = 400000  # Ancho de la imagen
img_height = 400000  # Alto de la imagen

# Añadir la rosa de los vientos en la nueva posición y tamaño
windrose_img = mpimg.imread('windrose.png')  # Asegúrate de tener una imagen llamada 'windrose.png' en tu directorio de trabajo
ax.imshow(windrose_img, extent=[img_x, img_x + img_width, img_y - img_height, img_y], zorder=5)

# Añadir el pie de página
fig.text(0.5, 0.05, "Fuente: Elaboración propia con datos del PREP 2024", ha='center', fontsize=12, color='#515B62')

# Añadir título y etiquetas
title_font = {'fontname': 'Chilanka', 'fontsize': 24, 'color': '#374EA2'}
ax.set_title('Resultados del PREP elección Federal', **title_font)

# Personalizar la leyenda
legend = ax.get_legend()
legend.set_bbox_to_anchor((0.25, 0.30))  # Posiciona la leyenda fuera del gráfico
legend.set_title('Candidatos presidenciales', prop={'size': 12})
plt.setp(legend.get_texts(), fontsize=12)  # Cambiar tamaño de fuente de los textos de la leyenda

# Mostrar el mapa
plt.show()
```
